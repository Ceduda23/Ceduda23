```python
import pandas as pd
from sklearn.model_selection import train_test_split
from keras import Sequential
from keras.layers import Dense
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
```

```python
dados = pd.read_csv('/content/diabetes.csv', sep=',')
```

```python
x = dados.values[:,0:8] # separação das características
y= dados.values[:,8] # separação da saida - 0 e 1
```

```python

x_treino, x_teste, y_treino, y_teste = train_test_split(x,y,test_size=0.20) # separação de 80% arquivo para treino
```

```python
model = Sequential() # criação do modelo
model.add(Dense(100, activation='tanh', kernel_initializer='he_normal', input_shape=(8,))) # camada de entrada
model.add(Dense(20,activation='relu')) # camada oculta
model.add(Dense(15,activation='relu')) # camada oculta
model.add(Dense(1,activation='sigmoid')) # camada de saída - usada em saída de classificação binária
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
history = model.fit(x_treino,y_treino,validation_data=(x_teste,y_teste), epochs=150,batch_size=32,verbose=1)
```

```python
plt.subplot(211)
#plt.title=('Grafico de custo') # Erros
plt.plot(history.history['loss'], label='Perdas durante o Treinamento')
plt.plot(history.history['val_loss'], label='Perdas durante os Testes')
plt.xlabel('Epocas')
plt.ylabel('Perdas')
plt.legend()
plt.show()
```
